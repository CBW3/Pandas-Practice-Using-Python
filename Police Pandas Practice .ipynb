```python
#loading in relevant libraries 
import pandas as pd 
```

```python
#loading & viewing the dataset 
police=pd.read_csv('Police.csv')
police.head()
```

```python
#basic EDA
police.shape
```

```python
police.dtypes
```

```python
#summary statistics table
police.describe()
```

```python
#obviously some nulls, going to dig in from here
police.isnull().sum()
```

```python
#jesus, ton of nulls, country name entire column is missing too 
#removing that colum
police.drop(columns= 'country_name', inplace=True)
```

```python
#checking to see if column is gone
police 
```

```python
#looks good, still gotta deal with other nulls 
police.isnull().sum()
```

```python
#creating new dataframe to drop nulls
df=police.dropna()
```

```python
#checking for nulls in newly created dataframe
df.isnull().sum()
```

```python
#looks good, taking a peek at cleaned data
df
```

```python
#going to see if men or women were stopped more often by the police for speeding violations
#gotta consider gender and violation columns respectively
df[df.violation == 'Speeding'].driver_gender.value_counts()
```

```python
#does gender affect who gets searched during a stop?
df.groupby('driver_gender').search_conducted.sum()
```

```python
#males are searched more often but also appear to be stopped more often as well 
```

```python
#finiding the mean stop duration
df.describe()
```

```python
#maybe stop duration isn't the right datatype?
df.dtypes
```

```python
#okay, so stop duration is essentially a categorical object in the dataset
#want to understand distribution first, then going to correct 
df.stop_duration.value_counts()
```

```python
#changing stop duration column
df['stop_duration']=df['stop_duration'].map({'0-15 Min':7.5, '16-30 Min':24, '30+ Min':45})
```

```python
#seeing if performed correctly
df
```

```python
#now, going to find mean stop duration
df.describe()
```

```python
#the mean stop duration is about 26 minutes give or take 
```

```python
#last, i want to compare the age distribtuion for each violation
df.groupby('violation').driver_age.describe()
```
